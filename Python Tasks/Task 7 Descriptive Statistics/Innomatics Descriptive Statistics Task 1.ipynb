# Different Statistical Measures 

## Measures of Center 
#### Our natural place to begin the analysis is to know about where the middle of the data is. 
#### In other words, what is the value that is closest to our other values..

## 1. Mean

### Mean is the most widely spread measure of Central tendency 
#### - It is calculated as simple average of the dataset
#### - Formula to calculate mean:
####                 Sum of all observations / total number of observations
#### - Note: It is easily affected by outliers

```python
from IPython.display import Image
Image(filename="C:/Innomatics/Mean.png",width=350)
```

```python
df['Mthly_HH_Income'].mean() #Average monthly household Income is 41,558/-
```

```python
df['Mthly_HH_Expense'].mean() #Average cost of monthly household Expences is 18,818/- 
```

```python
df['No_of_Earning_Members'].mean() #Average no of Members Earning in a family are 1 
                                   #probably 2 in some of the cases 
                                   #As mean is prone to outliers 
                                   #there may also be chances of 3 or 4 members in a very few cases and could be outliers
```

## 2. Median

#### -The data value that is positioned in the Middle of an oredered data set is referred as Median
#### - Also known as Middle most observation when the data is sorted in ascending/descending order.
#### - It is also said to be 50 percentile because 50% of the data is above it, and 50% below it.
#### - Formula for Median is the data value at position (n+1)/2 in an ordered dataset.
#### - If the position is not an integer, Median is given by the average of the data values closest to the calculated position.
#### - Note: Median is less impacted by the outliers.

```python
df['No_of_Earning_Members'].median() #Median No: of members earning in a family are 1
```

```python
df['No_of_Fly_Members'].median() #Median No: of members in a family are 4
```

```python
df['Mthly_HH_Income'].median() #Median of Monthly household income is 35,000/-
```

```python
df['Mthly_HH_Expense'].median() #Median of Monthly household Expence is 15,500/-
```

## 3.Mode 

#### - Mode is the data value that occurs most often.
#### - Observation which has highest frequency of occurence.
#### - It is also less impacted by outliers.
#### - It is very useful for categorical (non-numeric) data.
#### - Formula for Mode is calculated simply by finding the observation with highest frequency.
#### - Note: A dataset can have 0 modes, 1 mode or multiple modes.


```python
df['Emi_or_Rent_Amt'].mode() #Mode of Emi or Rent Ammount 
                             #Most frequent value occured of Emi or Rent Ammount is 0
```

```python
df['Highest_Qualified_Member'].mode() #Mode of Highest_Qualified_Member is Graduate
                                      #In most of the families Highest Qualified Member is a Graduate
```

```python
df['No_of_Earning_Members'].mode() #Mode of No: of members earning in a family is 1
                                   #In most of the families No: of earning members are only 1
```

## Measures of Dispersion or Spread
#### -   'Spread' refers loosely to how far away the more extreme values are from the center.
####  - In other words, Deviation of data values from the Mean

## 4. Variance
#### - It is the expectation of the squared deviation of a random variable from its mean.

#### - Population Variance: Average of squared deviations from the Mean
#### - In other words Variance tells us On average how far away all the data points are from the mean.
![title](Pictures/Variance.png)
## 5. Standard Deviation
#### - How close the values in a dataset are to the Mean
#### - Effectively measures the average distance away from the mean.
#### - Square root of the variance 
#### - There are different formulas for population and sample variance & standard deviation.
![title](Pictures/Std.png)

```python
df['No_of_Fly_Members'].var() #Average of squared deviations from the mean of No: of family members is 2.3
```

```python
df['No_of_Earning_Members'].var() #Average of squared deviations from the mean of No: of earning members in a family is 0.53
```

```python
df['No_of_Fly_Members'].std() #square root of variance of No: of family members 
```

```python
df['No_of_Earning_Members'].std() #square root of variance of No: of earning members in a family
```

## 6. Correlation
#### - Correlation is a measure of joint variability of two variables i.e. To find the strength of the relationship between 2 features or variables we use Correlation.
#### - It can be known as a standardized measure.
#### - It takes values between -1 and 1, making it easy for us to interpret the result.
#### - A correlation of 1, known as perfect positive correlation, means that one variable is perfectly explained by the other.
#### - A correlation of 0 means that the variables are independent.
#### -  A correlation of -1, known as perfect negative correlation, means that one variable is explaining the other one perfectly, but they move in opposite directions

## 7. Normal Distribution
#### • A normal distribution can be described with just two parameters, mean and standard deviation, given by the Greek mu (μ) and sigma (σ)
#### • It is represented as X ~ N(μ,σ²)
![title](Pictures/Normal_Distribution.png)
### 8. The normal distribution has several characteristics that make it very useful i.e.
#### • Symmetric around the mean
#### • Mean, median, mode are equal
#### • Area under the curve = 1
#### • Center of normal distribution curve gives average measurement of the data values
#### • How wide the curve is, is given by the standard deviation of the data values.
#### • Empirical rule: 68/95/99.7 
![title](Pictures/68-95-99.7Rule.jpg)

## 11. QQ Plot
#### - To Test if a distribution is Normal Distribution we use QQ plot i.e. Quantile Quantile plot
#### - While building any ML Algorithm like Linear Regression we often need to check if the residuals or errors (y_pred - y_actual) 
#### are normally distributed or not which is one of the assumptions of Linear Regression
#### - It helps us to know if any variable is normally distributed or not or is there any need to apply some transformation to the variable 
#### and then reattempt the modeling and then recheck the shape of the residuals.

```python
sns.histplot(df['Annual_HH_Income'],kde=True) 
df['Annual_HH_Income'].skew()         #Positively skewed with skewness value 1.19
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import statsmodels.api as sm
%matplotlib inline
sm.qqplot(df['Annual_HH_Income'], line='r');   #line = regression line
                                             #QQ plot of Annual_HH_Income variable before applying any kind of transformation
                                             #It can be observed that variable is not normally distributed 
                                             #As the histogram of Annual_HH_Income is slightly exponential (or) right skewed
                                             #Any type of transformation like log or boxcox
                                             #could be used to make it as normal distribution
```

```python
sns.histplot(np.log(df['Annual_HH_Income']+1),kde=True); #After applying log transformation 
                                               #it approximately looks as normal distribution though a bit negatively skewed
```

```python
transformed_var = np.log(df['Annual_HH_Income'])
transformed_var.skew()                           #skewness value after log transformation is -0.49
                                #it is almost converted to normal distribution but not completely normal distribution
                            #this could be achived using boxcox or any other transformation technique
```

```python
sm.qqplot(transformed_var, line='r'); #QQ plot of transformed variable Annual_HH_Income
                                     #We got a good overlap here between log transformed distribution quantiles 
                                     #and quantiles of a normal distribution
```

```python
from scipy import stats

df['Annual_HH_Income_boxcox'],l = stats.boxcox(df['Annual_HH_Income']+1) #If boxcox transformation is applied  

df['Annual_HH_Income_boxcox'].skew()                             #skewness value is decreased making it normal distribution
```

```python
sm.qqplot(df['Annual_HH_Income_boxcox'], line='r');   #QQ plot of transformed Annual_HH_Income using boxcox
                                                     #We got a better overlap here between boxcox transformed distribution 
                                                    #quantiles and quantiles of a normal distribution
```

## 12. Box-Cox Transformation
#### - To convert given distribution to normal distribution we use Box-Cox Transformation 
#### - Let X = [x1,x2,x3,….xn], then when we call box-cox(X), it returns a ‘lambda(λ)’ which we can use to generate Y as follows:

#### - For ‘i ‘ from [1,2,..n], do : 
####        yi = ((xi**λ)-1) / λ    if  λ  is  not  0
####        yi = log(xi)              if  λ  is  0

```python
from scipy import stats

df['Mthly_HH_Expense_boxcox'],lam = stats.boxcox(df['Mthly_HH_Expense']+1) #transforming training data & saving lambda value

print(lam)
```

```python
sns.histplot(df['Mthly_HH_Expense_boxcox'],kde=True)

df['Mthly_HH_Expense_boxcox'].skew()  #After applying transformation the skewness value came out to be -0.003 
                                      #Which is almost close to 0, indicating normal distribution
```

```python
df['Mthly_HH_Income_boxcox'],lam = stats.boxcox(df['Mthly_HH_Income']+1) #transforming training data & saving lambda value

print(lam)
```

```python
sns.histplot(df['Mthly_HH_Income_boxcox'],kde=True)

df['Mthly_HH_Income_boxcox'].skew() # Skewness value in this case aslo turns out to be closer to 0 
                                    #indicating normal distribution
```
